```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('baseball.csv')
```

```python
df.head()
```

```python
df['RD'] = df['RS'] - df['RA']
```

```python
df = df[df.Year < 2002]
```

```python
df.head()
```

```python
flatui = ["#6cdae7", "#fd3a4a", "#ffaa1d", "#ff23e5", "#34495e", "#2ecc71"]
```

```python
sns.set_palette(flatui)
sns.palplot(sns.color_palette())
```

```python
#blue indicates teams not selected in playoffs and red for team selected
```

```python
sns.lmplot(x='W',y='RD',fit_reg=False,hue='Playoffs',data=df,height=7, aspect=1.25) 
plt.xlabel('Wins',fontsize=20)
plt.ylabel('Run Difference',fontsize=20)
plt.axvline(99, 0, 1, color = "Black", ls = '--')
plt.show()
```

```python
#Relationship between Run Diff and Wins
```

```python
x = np.array(df.RD)
y = np.array(df.W)
slope, intercept = np.polyfit(x,y,1)
abline_values = [slope * i + intercept for i in x]
plt.figure(figsize=(10,8))
plt.scatter(x, y)
plt.plot(x, abline_values, 'r')
plt.title("Slope = %s" % (slope), fontsize = 12)
plt.xlabel("Run Difference", fontsize =20)
plt.ylabel("Wins", fontsize = 20)
plt.axhline(99, 0, 1, color = "k", ls = '--')
plt.show()
```
